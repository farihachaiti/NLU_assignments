```python
import nltk
from nltk.parse.generate import generate
from nltk import Nonterminal
from pcfg import PCFG

#2 sentences
sentences = ["I eat rice with fish", "She loves photography"]
#PCFG
wght_rules = [
    'S -> NP VP[1.0]',
    'VP -> V NP[0.7]',
    'VP -> V NP PP[0.15]',
    'VP -> PP [0.15]',
    'NP -> PRON [0.7]',
    'NP -> N [0.15]',
    'NP -> N PP [0.15]',
    'PP -> P NP [1.0]',
    'PRON -> "I" [0.7]',
    'PRON -> "She" [0.3]',
    'P -> "with" [1.0]',
    'N -> "rice" [0.33]',
    'N -> "fish" [0.33]',
    'N -> "photography" [0.34]',
    'V -> "eat" [0.6]',
    'V -> "loves" [0.4]' 
]
print("\033[1m2 Sentences:\033[0m")
for sent in sentences:   
    print(sent)
print("\033[1mPCFG Model:\033[0m")
for rule in wght_rules:
    print(rule)
# parsing
grammer = nltk.PCFG.fromstring(wght_rules)
parser = nltk.ViterbiParser(grammer)
print("\033[1mParsing of Sentences:\033[0m")
for sent in sentences:
    for tree in parser.parse(sent.split()):
        print(tree)

print("\033[1m10 sentences with nltk.parse.generate.generate using starting symbole 'S'\033[0m") 
start = Nonterminal('S')
for sent in generate(grammer, start=start, n=10):
    print(sent)

print("\033[1m10 sentences with nltk.parse.generate.generate using starting symbol 'V'\033[0m") 
start = Nonterminal('V')
for sent in generate(grammer, start=start, n=7):
    print(sent)
#10 sentences with PCFG.generate    
print("\033[1m10 sentences with PCFG.generate\033[0m") 
toy_grammar = PCFG.fromstring(wght_rules)
for sent in toy_grammar.generate(10):
    print(sent)
```
